```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# For reproducibility
pd.set_option('display.max_columns', None)
print("✅ Environment ready")
```

```python
# Core parameters
TICKER = "^NSEI"  # NIFTY 50 index on Yahoo Finance
START_DATE = "2010-01-01"  # ~15 years back
END_DATE = "2026-03-21"    # Today

print(f"📥 Downloading {TICKER} from {START_DATE} to {END_DATE}...")

# Download via yfinance
nifty_data = yf.download(TICKER, start=START_DATE, end=END_DATE, progress=False)

# Basic inspection
print(f"Shape: {nifty_data.shape}")
print("\nFirst 5 rows:")
nifty_data.head()
```

```python
# Save raw data to data/raw/
raw_path = "data/raw/nifty50_yahoo_raw.csv"
nifty_data.to_csv(raw_path)
print(f"💾 Raw data saved: {raw_path}")
print(f"Date range: {nifty_data.index.min()} to {nifty_data.index.max()}")
print(f"Trading days: {len(nifty_data)}")
```

```python
# Plot closing price
plt.figure(figsize=(12, 6))
plt.plot(nifty_data.index, nifty_data['Close'], linewidth=1)
plt.title('NIFTY 50 Closing Price (2010-2026)', fontsize=14, fontweight='bold')
plt.ylabel('Index Level')
plt.xlabel('Date')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("✅ Data pipeline complete!")
```
